```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

image = cv2.imread("frame_19.jpg", cv2.IMREAD_GRAYSCALE)

def apply_log_transformed(image):

    image2 = image.astype(np.float32)

    r_max = np.max(image2)

    c = 255 / np.log(1 + r_max)

    log_transformed = c * np.log(1 + image2)

    return np.array(log_transformed, dtype=np.uint8)

result = apply_log_transformed(image)

plt.figure(figsize=(10,5))

plt.subplot(1,2,1)
plt.title("Original Image")
plt.imshow(image, cmap='gray')

plt.subplot(1,2,2)
plt.title("Log Transformed")
plt.imshow(result, cmap='gray')

plt.show()
```
