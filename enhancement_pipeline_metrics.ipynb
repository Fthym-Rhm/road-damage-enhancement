```python
import cv2
import numpy as np
import math
from skimage.metrics import structural_similarity as ssim

original = cv2.imread("frame_19.jpg", cv2.IMREAD_GRAYSCALE)

def contrast_stretching(image):

    minval = np.min(image)
    maxval = np.max(image)

    dmin = 50
    dmax = 200

    stretched = dmin + (image - minval) * ((dmax - dmin) / (maxval - minval))

    return np.clip(stretched,0,255).astype(np.uint8)

enhanced = contrast_stretching(original)

mse = np.mean((original.astype("float") - enhanced.astype("float"))**2)

if mse == 0:
    psnr = 100
else:
    psnr = 20 * math.log10(255.0 / math.sqrt(mse))

ssim_value = ssim(original, enhanced, data_range=255)

print("MSE:", mse)
print("PSNR:", psnr)
print("SSIM:", ssim_value)
```
