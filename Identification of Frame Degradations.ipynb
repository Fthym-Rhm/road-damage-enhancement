```python
import cv2
import os

# Path to your video file
video_path = 'IMG_5985.mov'

# Folder to save extracted frames
output_folder = "extracted_frames"
os.makedirs(output_folder, exist_ok=True)

# Open video
cap = cv2.VideoCapture(video_path)

frame_count = 0
saved_count = 0

# Extract every Nth frame (adjust as needed)
frame_interval = 30   # Save 1 frame every 30 frames

while True:
    ret, frame = cap.read()
    if not ret:
        break

    if frame_count % frame_interval == 0:
        filename = os.path.join(output_folder, f"frame_{saved_count}.jpg")
        cv2.imwrite(filename, frame)
        saved_count += 1

    frame_count += 1

cap.release()
print("Frame extraction completed.")
```

```python
import cv2
import os

def detect_blur(image):
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    laplacian_var = cv2.Laplacian(gray, cv2.CV_64F).var()
    return laplacian_var

video_path =  'IMG_5985.mov'
output_folder = "blurry_frames"
os.makedirs(output_folder, exist_ok=True)

cap = cv2.VideoCapture(video_path)

threshold = 100  # Lower value = more blur sensitivity
frame_id = 0

while True:
    ret, frame = cap.read()
    if not ret:
        break

    blur_value = detect_blur(frame)

    if blur_value < threshold:
        cv2.imwrite(f"{output_folder}/blur_{frame_id}.jpg", frame)

    frame_id += 1

cap.release()
print("Blur detection completed.")
```

```python
import cv2
import numpy as np
import os

video_path ='IMG_5985.mov'
output_folder = "illumination_issues"
os.makedirs(output_folder, exist_ok=True)

cap = cv2.VideoCapture(video_path)
frame_id = 0

while True:
    ret, frame = cap.read()
    if not ret:
        break

    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    mean_intensity = np.mean(gray)

    # Low light detection
    if mean_intensity < 50:
        cv2.imwrite(f"{output_folder}/low_light_{frame_id}.jpg", frame)

    # Overexposed detection
    elif mean_intensity > 200:
        cv2.imwrite(f"{output_folder}/overexposed_{frame_id}.jpg", frame)

    frame_id += 1

cap.release()
print("Illumination analysis completed.")
```
